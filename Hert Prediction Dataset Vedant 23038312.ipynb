```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the Dataset
df = pd.read_csv('heart.csv')
```

```python
# Convert categorical variables to numerical

df['Sex'] = df['Sex'].map({'M': 0, 'F': 1})
df = pd.get_dummies(df, columns=['ChestPainType'])
```

```python
print(df.dtypes)
```

```python
#violinplot for resting ECG against Cholestrol

ax = sns.violinplot(x="RestingECG", y="Cholesterol", hue="Sex",data=df, palette="Set2", 
                    split=True,scale="count", inner="stick", scale_hue=False)
```

```python
import matplotlib.pyplot as plt

#Pie chart for different data type

# Get the value counts of data types
value_counts = df.dtypes.value_counts()

# Plotting
plt.figure(figsize=(6, 6))
matplotlib.rcParams.update({'font.size': 12})

# Determine the explode list length based on the number of unique data types
explode = [0.1] * len(value_counts)

# Plot the pie chart
value_counts.plot.pie(explode=explode, autopct='%1.2f%%', shadow=True)
plt.title('Data Type', color='Black', pad=20)
plt.axis('off')
plt.show()
```

```python
#Scatter plot of Resting BP vs Cholestrol level

# Create functions for plotting
def create_relational_graph():
    plt.figure(figsize=(8, 6))
    plt.scatter(df['RestingBP'], df['Cholesterol'])
    plt.xlabel('Resting Blood Pressure')
    plt.ylabel('Cholesterol Level')
    plt.title('Scatter Plot: Resting BP vs Cholesterol')
    plt.show()
create_relational_graph()
```

```python
#Bar chart of count of chest pain types

def create_categorical_graph():
    plt.figure(figsize=(8, 6))
    sns.countplot(x='ChestPainType_ATA', data=df)  # Choose one of the encoded columns
    plt.title('Count of Chest Pain Types')
    plt.xlabel('Chest Pain Type')
    plt.ylabel('Count')
    plt.show()
create_categorical_graph()
```

```python
#Heatmap for the complete data

# Create functions for plotting

def create_statistical_graph():
    # Select only numeric columns for correlation calculation
    numeric_columns = df.select_dtypes(include=['int64', 'float64'])
    
    # Calculate correlation matrix for numeric columns
    correlation_matrix = numeric_columns.corr()
    
    # Plot the correlation matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, vmax=.8, linewidths=0.01, annot=True, cmap='seismic', fmt='.2f',linecolor="black")
    plt.title('Correlation Matrix')
    plt.show()
create_statistical_graph()
```
